```python
from matplotlib.legend_handler import HandlerLine2D, HandlerTuple
```

```python
fig = plt.figure(figsize=[15,4])
plt.subplot(121)

m1, = plt.step(mat['Wp_simulated_in_vitro'][0],mat['Rho_Wp_simulated_in_vitro'][0], where='mid', 
         linewidth=2, color='darkcyan')
m2, = plt.step(mat['Wd_simulated_in_vitro'][0],mat['Rho_Wd_simulated_in_vitro'][0], where='mid', 
         linewidth=2, color='blue')
p1 = plt.scatter(mat['w_max_bin_in_vitro'],mat['Pw_max_in_vitro'], color='turquoise', s=60, 
            edgecolor='darkcyan', alpha=0.7)

p2 = plt.scatter(mat['w_min_bin_in_vitro'],mat['Pw_min_in_vitro'], color='deepskyblue', s=60,
           edgecolor='blue', alpha=0.7)

plt.xlim([0,10])
plt.ylim([0,3])

plt.legend([(m1, m2),(p1,p2)], ['Experiment', 'Simulation'], numpoints=1,
               handler_map={tuple: HandlerTuple(ndivide=None)}, fontsize=22)

plt.yticks(np.arange(0,4,1), fontsize=20)
plt.xticks(np.arange(0,12,2), fontsize=20)
ax = plt.gca()
ax.tick_params(width=2, length=4)
ax.spines['top'].set_linewidth(0)
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.spines['right'].set_linewidth(0)

plt.xlabel('$I/K_{0}$', fontsize=25)
plt.ylabel('$P(I/K_{0})$', fontsize=25)



plt.subplot(122)

m1, = plt.step(mat['Tp_simulated_in_vitro'][0],mat['Rho_Tp_simulated_in_vitro'][0], where='mid', 
         linewidth=2, color='mediumvioletred', label='Simulation')
m2, = plt.step(mat['Td_simulated_in_vitro'][0],mat['Rho_Td_simulated_in_vitro'][0], where='mid', 
         linewidth=2, color='darkorange', label='Simulation')
p1 = plt.scatter(mat['T_max_bin_in_vitro'],mat['PT_max_in_vitro'], color='hotpink', s=60, 
            edgecolor='mediumvioletred', alpha=0.7,  label='Experiment')
p2 = plt.scatter(mat['T_min_bin_in_vitro'],mat['PT_min_in_vitro'], color='bisque', s=60, 
            edgecolor='darkorange', alpha=0.7, label='Experiment')

plt.xlim([0,4])
plt.ylim([0,2])
plt.xticks(np.arange(0,5,1))
plt.yticks(np.arange(0,2.5,0.5))

plt.legend([(m1, m2),(p1,p2)], ['Experiment', 'Simulation'], numpoints=1,
               handler_map={tuple: HandlerTuple(ndivide=None)}, fontsize=22)


ax = plt.gca()
ax.tick_params(width=2, length=4)
ax.spines['top'].set_linewidth(0)
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.spines['right'].set_linewidth(0)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel(r'$Time/T_{0}$', fontsize=25)
plt.ylabel(r'$P(Time/T_{0})$', fontsize=25)






plt.show()

fig.savefig('../Data/Figure 4/Figure 4 - simulations/Results/Distributions-InVitro.pdf', 
           bbox_inches='tight')
```
